# Perceptron Example - Breast Cancer

In this notebook the Perception module from the rice_ml package is used to classify tumor cell data.
Prediction: Malignant or Benign tumor, binary encoded: 1, -1


```python
import numpy as np
import pandas as pd
#import sys
#import os
#sys.path.append(os.path.abspath("../../../src"))

from rice_ml.supervised_learning import Perceptron

import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
```

# Dataset Loading

We use the Breast Cancer data set from UCI. The dataset includes 569 samples and 32 features - ID, Diagnosis, then 3 sets of the following:
- radius
- texture
- perimeter
- area
- smoothness
- compactness
- concavity
- concave points
- symmetry
- fractal dimension

The data is loaded into a pandas dataframe.

```python
input_file = "wdbc.data"
columns = [
    'ID', 'Diagnosis',
    'radius1', 'texture1', 'perimeter1', 'area1', 'smoothness1',
    'compactness1', 'concavity1', 'concave_points1', 'symmetry1', 'fractal_dimension1',
    
    'radius2', 'texture2', 'perimeter2', 'area2', 'smoothness2',
    'compactness2', 'concavity2', 'concave_points2', 'symmetry2', 'fractal_dimension2',
    
    'radius3', 'texture3', 'perimeter3', 'area3', 'smoothness3',
    'compactness3', 'concavity3', 'concave_points3', 'symmetry3', 'fractal_dimension3'
]
df = pd.read_csv(input_file, header=None, names=columns)

df
```

# Dataset Preprocessing

The data is then placed into feature matrix `X` and label vector `y`. The ID feature is removed from the dataset, as it is not useful for predictions. The sci-kit scaler is applied in order to standardize feature magnitude.

```python
X = df.drop(['Diagnosis', "ID"], axis=1)
X = df[['radius1', 'texture1', 'perimeter1', 'area1', 'smoothness1',
    'compactness1', 'concavity1', 'concave_points1', 'symmetry1', 'fractal_dimension1',
    
    'radius2', 'texture2', 'perimeter2', 'area2', 'smoothness2',
    'compactness2', 'concavity2', 'concave_points2', 'symmetry2', 'fractal_dimension2',
    
    'radius3', 'texture3', 'perimeter3', 'area3', 'smoothness3',
    'compactness3', 'concavity3', 'concave_points3', 'symmetry3', 'fractal_dimension3']] 
y = df['Diagnosis']


y = y.replace({"M": 1, "B": -1})

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

The data is split between training and testing data at an 80/20 ratio.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=25, stratify=y
)
```

# Model Training

The model is initialized with a learning rate of 0.01 and a maximum of 1000 iteration, and trained using the split off training data. 

```python
model = Perceptron(0.01, 1000)
model.train(X_train, y_train)
```

# Prediction and Evaluation
The model makes predictions on the remaining testing data, and is evaluated for accuracy.  
The confusion matrix shows the following \(Malignant = Positive\):  
\[True Positive   False Positive\]  
\[False Negative   True Negative\]
The various weights of the 30 remaining features are also shown.

```python
y_pred = model.predict(X_test)
y_test = y_test.astype(int)
accuracy = accuracy_score(y_test, y_pred)
print(f"Test Accuracy: {accuracy:.2%}")

# Confusion matrix
cm = confusion_matrix(y_test, y_pred, labels=[1, -1])
print("Confusion Matrix:\n", cm)

print("Weights:", model.w_)
```

 # Convergence
 A graph shows the number of errors occuring over each iteration of the model.

```python
plt.plot(range(1, len(model.errors_) + 1), model.errors_, marker='o')
plt.xlabel('Iterations')
plt.ylabel('Number of Misclassifications')
plt.title('Perceptron Training Error Over Time')
plt.show()
```
